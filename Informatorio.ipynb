# **VISUALIZACION DE DATOS**

Para el siguiente analisis tomamos como referencia los nacimientos ocurridos en Argentina durante los aÃ±os del 2012 al 2022. Tambien agregamos datos del tipo de sexo ocurridos en esos nacimientos.

**Grafico 1:** En el siguiente grafico podemos apreciar la cantidad de nacimientos durante las fechas mencionadas. Podemos ver que del 2012 al 2014 un leve acenso en los nacimientos y a partir de esa fecha una decrecimiento considerable.

```python
df = births.groupby('anio').sum()
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

sns.lineplot(data=df, x="anio", y='nacimientos_cantidad',palette='crest')
plt.xlabel('AÃ±o')
plt.xticks(ticks=range(2012,2023,1))
plt.yticks(ticks=range(440000,800000,20000))
plt.ylabel('Nacimientos')
plt.title('Nacimientos por AÃ±o')
plt.show()

```

**Grafico 2:**En el siguiente grafico podemos apreciar en los nacimiento que genero en cuention se le atribuyo. Podemos apreciar que en el rango de aÃ±os del 2012 al 2022 el sexo masculumo tuvo 4.000.000 nacimiento contra aproximadamente 3.700.000, que corresponden al sexo masculino 52,4% y el sexo femenino al 47,6%

```python
anios = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
data = names_frequency[names_frequency["anio"].isin(anios)]

names =  data.groupby("sexo").sum()

sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.barplot(data=names, x="sexo", y='cantidad', palette="flare")
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.title('Cantidad de nacimientos por sexo')
plt.show()

names = names_frequency.groupby("sexo").sum()
total_sexo = names_frequency["cantidad"].sum()
names["porcentaje"] = (names["cantidad"] / total_sexo) * 100

plt.pie(names["porcentaje"].values, labels=names.index, autopct='%1.1f%%')
plt.title('Porcentaje de Nacimientos por Sexo')
plt.show()
```

**Grafico 3:**En el siguiente grafico muestra un desgloce de los aÃ±os y que genero predomino en esos aÃ±os. Podemos ver que en los primeros aÃ±os se ve un marcado numero de nacimientos del sexo masculino y este va decreciendo hasta llegar al 2019 donde este se iguala, los aÃ±os siguientes el sexo femenino va superando al otro sexo en cuestion.

```python

anios = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
data = names_frequency[names_frequency["anio"].isin(anios)]

data =  data.groupby(["sexo","anio"]).sum()

sns.barplot(data=data, x="anio", y='cantidad', hue='sexo')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad')
plt.title('Cantidad de Nacimientos por AÃ±o y Sexo')
plt.show()
```

**Grafico 4:**En el siguiente grafico podemos observar un desgloce de los nacimientos entre 2012 al 2022 en las diferente provincias que comprende la Argentina. Podemos observar que, como el grafico 2 se ve un aumento y un decenso en los nacimientos en la mayoria de las de las provincias. Las 3 provincias que tuvieron la mayoria de nacimientos fueron Buenos Aires, Cordoba y Santa Fe.

```python

tabla = births.pivot_table(index='provincia_nombre', columns='anio', values='nacimientos_cantidad', fill_value=0)

plt.figure(figsize=(20, 10))
sns.heatmap(tabla, cmap="YlOrBr", annot=True, fmt=".0f")
plt.title("Nacimientos por Provincia y AÃ±o")
plt.xlabel("AÃ±o")
plt.ylabel("Provincia")
plt.tight_layout()
plt.show()
```

**Grafico 5:**En el siguiente grafico podemos observar que porcentaje en los aÃ±os tomados en los datos equivalen los nacimientos en cada provincia. Predominando notablemente en la provincia de Buenos Aires

```python
nacimientos_cantidad_total = births['nacimientos_cantidad'].sum()
df = births.groupby("provincia_nombre").sum()
df['porcentaje'] = (df['nacimientos_cantidad'] / nacimientos_cantidad_total) * 100
data = df.sort_values(by='porcentaje', ascending=False )

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

sns.barplot(data=data, x="provincia_nombre", y='porcentaje',palette='crest')
plt.xlabel('Provincia')
plt.ylabel('Porcentaje de nacimiento')
plt.yticks(ticks=range(0,40,2))
plt.xticks(rotation=90)
plt.title('Porcentaje de Nacimientos por Provincia')
plt.show()
```

**Grafico 6:**En el siguiente grafico podemos observar el total de los nacimientos entre el 2012 y 2022 por deparatamentos en la provincia del chaco, viendo que los 3 departamentos que poseen la mayor cantidad son San Fernando, General Guemes y Comandante Fernadez.

```python
df = births[births["provincia_nombre"] == "Chaco"]
df = df.groupby(["departamento_nombre","anio"]).sum().sort_values(by='nacimientos_cantidad', ascending=False)

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.barplot(data=df, x="departamento_nombre", y='nacimientos_cantidad',palette='crest')
plt.xlabel('Departamento')
plt.ylabel('Nacimientos')
plt.xticks(rotation=90)
plt.title('Nacimientos por Departamento en Chaco entre el 2012 y 2022')
plt.show()
```

**Grafico 6:**En el siguiente grafico podemos observar la ditribucion de sexo en nacidos entre el 2012 y 2022 por provincias

```python
data = names_frequency.groupby(["sexo","nombre_provincia"]).sum()

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.barplot(data=data, x="nombre_provincia", y='cantidad', hue='sexo')
plt.xlabel('Provincia')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.title('Cantidad de Nacimientos por Provincia y Sexo')
plt.show()
```

**Grafico 8:**En el siguiente grafico podemos observar una correlacion bastante alta entre los primeros aÃ±os, tambien podemos observar que los aÃ±os estan relacionados entre si, los datos de cada aÃ±o se asemejan.

```python
pivot = births.pivot_table(index='provincia_nombre', columns='anio', values='nacimientos_cantidad', fill_value=0)
correlatividad = pivot.corr()

plt.figure(figsize=(12, 6))
sns.heatmap(correlatividad, annot=True, cmap='YlOrBr')
plt.show()
```

# CONCLUSION





En el presente analisis de los datos podemos ver las siguientes cosas


1. Se muestra un alza de nacimientos hasta el 2014 y posteriormente un decenso significativo en los nacimientos.
2.   Este desenso y crecimiento se ve tanto a nivel nacional como por provincia individualmente.
3.   Tambien se analizÃ³ el genero de cada nacimiento en a nivel total reflejando mas nacimientos del genero masculino que del femenino pero con poca diferencia entre ambos.
4. Se observo que en el rango de aÃ±os analizados hay una tendencia decreciente de nacimientos del sexo masculino y llegando al 2019 este se igual con el femenino. A partir de ese aÃ±o empiza a predominar el femenino sobre el masculino.
5. Desglozando por provincia en general vemos que el genero masculono predomina sobre el femenido en la mayoria de las provincias  y en solo 4 provincias esto no es asi.



Podemos llegar a las siguientes hipotesis con respecto a los datos obtenidos.
1. El decenso de nacimientos puede esta ligado a:
    *. Situacion economica de pais, la personas concideran que criar a un hijo es costoso o piensan que no podran darle todos los gastos que requieran
    *. Cambios en el pensamiento de las personas, las pesonas poseen menor presion social para tener un hijo y priorizan su desarrollo personal antes que formar una familia.
    *. Disponibiliad de metodos anticonceptivos, las personas usa estos tantos para prevenir enfermedades y a su vez evitar la concepcion de un hijo.

2. Con respecto a los diferentes sexos entre los nacidos podemos decir que el decenso que se persive en los graficos puede estar ligado a cambios geneticos debido a las mezclas de genes en diferentes generaciones.
